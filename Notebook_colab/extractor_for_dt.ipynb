# For the Train Dataset    

## Libraries
It is necessary to import some libraries that allow specific operations:
  1. Import the `drive`module to access Google Drive from Google Colab and use it as storage;
  2. Import the `pandas`library for tabular data manipulation and **DataFrame** management;
  3. Import the `files`module to allow downloading the file from Colab to the local computer.

```python
from google.colab import drive  # [1]
import pandas as pd  # [2]
from google.colab import files  # [3]
```

## Mounting Drive
A window is opened to perforM **Google authentication** and access files from Colab.<br>Mount Drive in the`/content/drive`directory.

```python
drive.mount('/content/drive')
```

## Defining the path and the files to load
  1. The folder containing the files to process is specified through a given `base_path`;
  2. A list of `.parquet` files to be merged is created; these are partitions of the same TRAIN dataset under consideration (https://huggingface.co/datasets/Jinyan1/COLING_2025_MGT_multingual/viewer/default/train).

```python
# --------------- CHANGE THE PATH ----------------
base_path = "/content/drive/MyDrive/TESI - Classificazione/MIO/partizioni_dt"  # [1]
paths = [
    f"{base_path}/train-00000-of-00003.parquet",
    f"{base_path}/train-00001-of-00003.parquet",
    f"{base_path}/train-00002-of-00003.parquet"
]  # [2]
```

## Loading and merging the files
  1. Each  `.parquet` file is loaded into a separate **DataFrame**;
  2. All DataFrames are merged into a _single one_ (`df_full`).<br> `ignore_index=True` ensures that the index is regenerated from scratch.

```python
print("Loading Parquet files...")  # Message for the user
df_list = [pd.read_parquet(path) for path in paths]  # [1]: Read each file
df_full = pd.concat(df_list, ignore_index=True)  # [2]: Merge all DataFrames
```

## Displaying languages and user input
  1. Extraction of all unique values from the `"lang"` column, ignoring any NaN (_Not a Number_) values;
  2. The languages present in the dataset are printed.
  3. The user is asked to choose a **valid language among those listed**, repeating the prompt until a correct input is provided.

```python
available_languages = df_full["lang"].dropna().unique().tolist()  # [1]
print("Languages available in the dataset:")
print(available_languages)  # [2]

# [3]:
selected_language = ""
while selected_language not in available_languages:
    selected_language = input("Enter one of the available languages (e.g. 'it', 'en', etc.): ").strip()
    if selected_language not in available_languages:
        print("Invalid language. Please try again.")
```

## DataFrame filtering
  1. Creation of a new DataFrame containing _ONLY THE ROWS IN THE SELECTED LANGUAGE_;
  2. The number of filtered rows is printed.

```python
print(f"Filtering language: {selected_language}...")
df_lang = df_full[df_full["lang"] == selected_language]  # [1]
print(f"Total rows for language '{selected_language}': {len(df_lang)}")  # [2]
```

## Saving and Download
  1. The path and filename of the `.csv` file to save are constructed, including the language in the name;
  2. The filtered DataFrame is saved in CSV format on Google Drive. <br> `index=False` prevents the index column from being saved;
  3. The CSV file download to the user's computer is started.

```python
# ----------------- CHANGE THE PATH ----------------
output_csv = f"/content/drive/MyDrive/TESI - Classificazione/MIO/{selected_language}_train_full.csv"  # [1]
df_lang.to_csv(output_csv, index=False)  # [2]
print(f"CSV file saved as: {output_csv}")  # Message for the user
files.download(output_csv)  # [3]
```
